```python
import requests
import re

url = "http://127.0.0.1:7230/"
```

```python
# start session
data = {
    "session_id": "1",
    "remote_env_info": None,
    "name": "start_session",
    "arguments": {},
}

try:
    resp = requests.post(url, json=data)
    print("Status Code:", resp.status_code)
    print("Response:\n")
    print(resp.json()["output"])
except Exception as e:
    print(resp.text)
    print("Error:", e)
```

```python
# search
data = {
    "session_id": "1",
    "remote_env_info": {"search_forbidden_strs": ["Are the Laleli Mosque and Esma Sultan Mansion located in the same neighborhood?"]},
    "name": "browser.search",
    "arguments": {"query": "Are the Laleli Mosque and Esma Sultan Mansion located in the same neighborhood?"},
}

try:
    resp = requests.post(url, json=data)
    print("Status Code:", resp.status_code)
    print("Response:\n")
    print(resp.json()["output"])
except Exception as e:
    print(resp.text)
    print("Error:", e)
```
